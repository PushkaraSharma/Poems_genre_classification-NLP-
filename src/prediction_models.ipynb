```python
import pandas as pd
import numpy as np
```

```python
#loading the csv file
```

```python
df = pd.read_csv('final_poem_data.csv')
```

```python
df.head()
```

```python
x = df['content'].values
y = df['type'].values
```

```python
import string
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem import PorterStemmer
stop_words = stopwords.words('english')
ps = PorterStemmer()
```

```python
def preprocessing(words):
   # words = words.encode('ascii','ignore')
    words = str(words)
    words = words.lower()
    processed = words.replace(r'$|\£|\€', 'money')
    processed = processed.replace(r'\d+(\.\d+)?', 'number')
    processed = processed.replace(r'^\s+|\s+?$', '')
    processed = processed.replace(r'\s+', ' ')
    processed = processed.replace(string.punctuation,'')
    processed = processed.replace('=','')
    processed = processed.replace('\r','')
    processed = processed.replace('\n','')
    processed = processed.replace("?",'')
    processed = processed.replace(',','')
    processed = processed.replace("'",'')
    processed = processed.replace('|','')
    processed = processed.replace('.','')
    processed = processed.replace('!','')
    processed = processed.replace('_',' ')
    processed = processed.replace('-',' ')
    processed = processed.replace('"','')
    
    return processed

def bag_words(processed_words):
    tokens = word_tokenize(processed_words)
    clean = []
    for i in tokens:
        i = ps.stem(i)
        if i not in stop_words and i not in string.punctuation and i.startswith('//')==False:
            clean.append(i)
    dicton = dict([word, True] for word in clean)
    return dicton
        
```

```python
final = []
for i in range(len(x)):
    final.append(preprocessing(x[i]))
final_2 = []
for i in range(len(x)):
    final_2.append((bag_words(final[i]),y[i]))
```

```python
final_2
```

```python
from sklearn.model_selection import train_test_split
training,testing = train_test_split(final_2,test_size = 0.20)
```

```python
from nltk import NaiveBayesClassifier
from nltk import classify
```

## Model 1-Naive Bayes 

```python
classifier = NaiveBayesClassifier.train(training)
acc = classify.accuracy(classifier,testing)
acc
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
tf = TfidfVectorizer(stop_words=stop_words,max_features=8000)
X = tf.fit_transform(final)
```

```python
X.shape
```

## Model 2 - Logistic Regression

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
log = LogisticRegression()
log.fit(x_train, y_train)
```

```python
pred_y1 = log.predict(x_test)
accuracy_score(y_test,pred_y1)
```

## Model 3 - SVM(Linear)

```python
svm = SVC(kernel='linear')
svm.fit(x_train,y_train)
```

```python
pred_y2 = svm.predict(x_test)
accuracy_score(y_test,pred_y2)
```

## Model 5 - Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
tr = DecisionTreeClassifier()
tr.fit(x_train,y_train)
pred_y4 = tr.predict(x_test)
accuracy_score(y_test,pred_y4)
```
