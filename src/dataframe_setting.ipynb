```python
df1 = pd.read_csv('all.csv')
```

```python
df1 = df1.drop(['author','age'],axis=1)
```

```python
df1 = df1[['poem name','content','type']]
df1.head()
```

```python
df1['type'].value_counts()
```

```python
df2 = pd.read_csv('poem_data.csv')
```

```python
df2.columns = ['poem name','content','type']
df2.head()
```

```python
df2['type'].value_counts()
```

```python
df3 = pd.read_csv('poem_data_methology.csv')
```

```python
df3.columns = ['poem name','content','type']
df3 = df3.dropna()
df3.head()
```

```python
df4 = pd.concat([df1,df2,df3])
df4.shape
```
